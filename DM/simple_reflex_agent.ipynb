# Simple Reflex Agent

```python
def simple_reflex_agent(percept):
    location, status=percept
    
    if status=="Dirty":
        return "Suck"
    elif location=="A":
        return "Right"
    elif location=="B":
        return "Left"
    
def run_agent():
    environment=[("A","Dirty"),("A","Clean"),("B","Dirty"),("B","Clean")]
    
    for percept in environment:
        action= simple_reflex_agent(percept)
        print(f"The percept: {percept}-> Action: {action}")
```

```python
if __name__=="__main__":
    run_agent()
```

```python
def simple(percept):
    location,status=percept
    
    if status=="dirty":
        return "suck"
    elif location=="a":
        return "right"
    elif location=="b":
        return "left"
    
def run_agent():
    environment=[("a","dirty"),("a","clean"),("b","dirty"),("b","clean")]
    
    for percept in environment:
        action=simple(percept)
        print(f"The percept: {percept}->Action: {action}")
        
if __name__=="__main__":
    run_agent()
```
