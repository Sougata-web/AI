# Goal Based Agent

```python
class GoalBasedAgent:
    def __init__(self):
        self.model={
            "A":None,
            "B":None
        }
        self.location="A"
        self.goal={"A":"Clean","B":"Clean"}
        
    def update_model(self,percept):
        location,status=percept
        self.location=location
        self.model[location]=status
        
    def is_goal_achieved(self):
        return self.model==self.goal
    
    def formulate_goal(self):
        return self.goal
    
    def formulate_problem(self):
        actions=[]
        
        if self.model[self.location]=="Dirty":
            actions.append("Suck")
        elif self.location=="A":
            actions.append("Right")
        elif self.location=="B":
            actions.append("Left")
        return actions
    
    def choose_action(self,percept):
        self.update_model(percept)
        if self.is_goal_achieved():
            return "No Op"
        actions=self.formulate_problem()
        return actions[0]
    
    def run(self, environment): 
        for percept in environment: 
            print(f"Percept: {percept}\n") 
            action = self.choose_action(percept) 
            print(f"Action: {action}\n") 
            print(f"Internal Model: {self.model}\n") 
            print(f"Goal State: {self.goal}\n") 
            print("----------\n")  
        
        
```

```python
if __name__=="__main__":
    environment = [("A", "Dirty"), ("B", "Dirty"), ("A", "Clean"), ("B", "Clean")] 
 
agent = GoalBasedAgent() 
agent.run(environment) 
```
