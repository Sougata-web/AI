# Model Based Agent

```python
class ModelBasedReflexAgent:
    def __init__(self):
        self.model={
            "A":None,
            "B":None
        }
        self.location="A"
        
    def update_model(self,percept):
        location,status=percept
        self.location=location
        self.model[location]=status
        
    def decide_action(self): 
        location = self.location 
        status = self.model[location]
        
        if status=="Dirty":
            return "Suck"
        elif location=="A":
            return "Left"
        elif location=="B":
            return "Right"
        
    def run(self,environment):
        for percept in environment:
            print(f"Percept: {percept}") 
            self.update_model(percept) 
            action = self.decide_action() 
            print(f"Action: {action}") 
            print(f"Internal Model: {self.model}") 
            print("----------")
```

```python
if __name__=="__main__":
    environment_percepts = [("A", "Dirty"), ("B", "Dirty"), ("A", "Clean"), ("B", "Clean")] 
    agent=ModelBasedReflexAgent()
    agent.run(environment_percepts)
```
